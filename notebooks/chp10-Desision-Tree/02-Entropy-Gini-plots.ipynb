# 信息熵

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2分类信息熵
def entropy(p):
    return -p * np.log(p) - (1-p) * np.log(1-p)
```

```python
x = np.linspace(0.01, 0.99, 200)
```

```python
plt.plot(x, entropy(x))
plt.show()
```

```python
def gini(x):
    return -2 * (x ** 2) + 2 * x
```

```python
plt.plot(x, gini(x))
plt.show()
```
